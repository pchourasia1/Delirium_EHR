```python
from sklearn.metrics import average_precision_score, precision_recall_curve

def roc_auc_score_multiclass(actual_class, pred_class, average="macro"):
    unique_class = set(actual_class)
    roc_auc_dict = {}
    for per_class in unique_class:
        other_class = [x for x in unique_class if x != per_class]
        new_actual_class = [0 if x in other_class else 1 for x in actual_class]
        new_pred_class = [0 if x in other_class else 1 for x in pred_class]
        roc_auc = roc_auc_score(new_actual_class, new_pred_class, average=average)
        roc_auc_dict[per_class] = roc_auc
    check = pd.DataFrame(roc_auc_dict.items())
    return mean(check)


##########################  SVM Classifier  ################################
def svm_fun(X_train, y_train, X_test, y_test):
    X_train = preprocessing.scale(X_train)
    X_test = preprocessing.scale(X_test)

    start = timeit.default_timer()
    clf = svm.SVC(kernel='linear', probability=True, class_weight='balanced')
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    svm_acc = metrics.accuracy_score(y_test, y_pred)
    svm_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    svm_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    svm_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    svm_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = clf.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [svm_acc, svm_prec, svm_recall, svm_f1_weighted, svm_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  NB Classifier  ################################
def gaus_nb_fun(X_train, y_train, X_test, y_test):

    start = timeit.default_timer()
    gnb = GaussianNB()
    y_pred = gnb.fit(X_train, y_train).predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    NB_acc = metrics.accuracy_score(y_test, y_pred)
    NB_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    NB_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    NB_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    NB_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = gnb.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [NB_acc, NB_prec, NB_recall, NB_f1_weighted, NB_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  MLP Classifier  ################################
def mlp_fun(X_train, y_train, X_test, y_test):

    start = timeit.default_timer()
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test_2 = scaler.transform(X_test)

    mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)
    mlp.fit(X_train, y_train)
    y_pred = mlp.predict(X_test_2)
    stop = timeit.default_timer()
    time_new = stop - start

    MLP_acc = metrics.accuracy_score(y_test, y_pred)
    MLP_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    MLP_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    MLP_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    MLP_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = mlp.predict_proba(X_test_2)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [MLP_acc, MLP_prec, MLP_recall, MLP_f1_weighted, MLP_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  KNN Classifier  ################################
def knn_fun(X_train, y_train, X_test, y_test):

    start = timeit.default_timer()
    classifier = KNeighborsClassifier(n_neighbors=5)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    knn_acc = metrics.accuracy_score(y_test, y_pred)
    knn_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    knn_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    knn_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    knn_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = classifier.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [knn_acc, knn_prec, knn_recall, knn_f1_weighted, knn_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  Random Forest Classifier  ################################
def rf_fun(X_train, y_train, X_test, y_test):
    from sklearn.ensemble import RandomForestClassifier
    start = timeit.default_timer()
    rf = RandomForestClassifier(n_estimators=100, class_weight='balanced')
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    fr_acc = metrics.accuracy_score(y_test, y_pred)
    fr_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    fr_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    fr_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    fr_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = rf.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [fr_acc, fr_prec, fr_recall, fr_f1_weighted, fr_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  Logistic Regression Classifier  ################################
def lr_fun(X_train, y_train, X_test, y_test):
    X_train = preprocessing.scale(X_train)
    X_test = preprocessing.scale(X_test)

    start = timeit.default_timer()
    model = LogisticRegression(solver='liblinear', random_state=0,
                               class_weight='balanced')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    LR_acc = metrics.accuracy_score(y_test, y_pred)
    LR_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    LR_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    LR_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    LR_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = model.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [LR_acc, LR_prec, LR_recall, LR_f1_weighted, LR_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  Decision Tree Classifier  ################################
def fun_decision_tree(X_train, y_train, X_test, y_test):
    from sklearn import tree
    start = timeit.default_timer()
    clf = tree.DecisionTreeClassifier(class_weight='balanced')
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    dt_acc = metrics.accuracy_score(y_test, y_pred)
    dt_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    dt_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    dt_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    dt_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = clf.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [dt_acc, dt_prec, dt_recall, dt_f1_weighted, dt_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  Balanced Random Forest Classifier  ################################
def brf_fun(X_train, y_train, X_test, y_test):
    from imblearn.ensemble import BalancedRandomForestClassifier
    start = timeit.default_timer()
    brf = BalancedRandomForestClassifier(n_estimators=100, sampling_strategy='all',
                                         replacement=True, bootstrap=False, random_state=0)
    brf.fit(X_train, y_train)
    y_pred = brf.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    brf_acc = metrics.accuracy_score(y_test, y_pred)
    brf_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    brf_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    brf_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    brf_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = brf.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [brf_acc, brf_prec, brf_recall, brf_f1_weighted, brf_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check


##########################  XGBoost Classifier  ################################
def xgb_fun(X_train, y_train, X_test, y_test):
    from xgboost import XGBClassifier
    scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

    start = timeit.default_timer()
    xgb = XGBClassifier(scale_pos_weight=scale_pos_weight, eval_metric='logloss', random_state=0)
    xgb.fit(X_train, y_train)
    y_pred = xgb.predict(X_test)
    stop = timeit.default_timer()
    time_new = stop - start

    xgb_acc = metrics.accuracy_score(y_test, y_pred)
    xgb_prec = metrics.precision_score(y_test, y_pred, average='weighted')
    xgb_recall = metrics.recall_score(y_test, y_pred, average='weighted')
    xgb_f1_weighted = metrics.f1_score(y_test, y_pred, average='weighted')
    xgb_f1_macro = metrics.f1_score(y_test, y_pred, average='macro')

    y_prob = xgb.predict_proba(X_test)[:, 1]
    macro_roc_auc_ovo = roc_auc_score(y_test, y_prob)
    auprc = average_precision_score(y_test, y_prob)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    check = [xgb_acc, xgb_prec, xgb_recall, xgb_f1_weighted, xgb_f1_macro,
             macro_roc_auc_ovo, auprc, bal_acc, mcc, time_new]
    return check
```

## **4.1 · Cap Outliers** (F→C conversion + clip vitals to clinical ranges)


```python
#Cap Outliers
import pandas as pd

df = pd.read_csv("cohort_all_features_final.csv")
print(f"Loaded: {df.shape}")

# Convert temperature_last from Fahrenheit to Celsius
df["temperature_last"] = (df["temperature_last"] - 32) * 5 / 9
print("Converted temperature_last from F to C")

# Define caps: column prefix -> (lower, upper)
CAPS = {
    "heart_rate": (25, 225),
    "resp_rate": (4, 60),
    "sbp": (40, 260),
    "dbp": (20, 160),
    "mbp": (25, 200),
    "temperature": (30, 45),
}

SUFFIXES = ["_min", "_max", "_mean", "_last"]

for prefix, (lo, hi) in CAPS.items():
    for suffix in SUFFIXES:
        col = prefix + suffix
        if col not in df.columns:
            continue
        before = df[col].describe()
        clipped = df[col].clip(lower=lo, upper=hi)
        n_lo = (df[col] < lo).sum()
        n_hi = (df[col] > hi).sum()
        df[col] = clipped
        if n_lo > 0 or n_hi > 0:
            print(f"  {col}: capped {n_lo} low, {n_hi} high")

df.to_csv("cohort_all_features_capped.csv", index=False)
print(f"\nSaved: cohort_all_features_capped.csv")
```

## **4.3 · Encode Binary Columns (0=false, 1=true, 2=unknown)**

## We need to Deal/Impute Missing Values

```python
import pandas as pd

df = pd.read_csv("cohort_all_features_capped.csv")
print(f"Loaded: {df.shape}")

# ------------------------------------------------------------------
# Drop columns with over 40% missingness (unchanged)
# ------------------------------------------------------------------
threshold = 0.40  # drop columns missing more than this fraction
missing_frac = df.isna().mean()
to_drop = missing_frac[missing_frac > threshold].index.tolist()

df = df.drop(columns=to_drop)
print(f"Dropped {len(to_drop)} columns over {threshold:.0%} missing:")
for col in to_drop:
    print(f"  {col} ({missing_frac[col]:.1%})")

# ------------------------------------------------------------------
# Categorical encoding of binary columns
#   0 = false / absent
#   1 = true  / present
#   2 = unknown / missing
#
# A column is treated as binary if all of its non-null values are a
# subset of {0, 1}. Only binary columns that actually have missing
# values are re-encoded. Detection is done from the data itself, so
# there is no hand-kept list that can drift out of sync with the
# feature set. Columns already stored as {0,1,2} (e.g. on a re-run)
# are skipped because their value set is no longer a subset of {0,1}.
# ------------------------------------------------------------------

# Never encode these, even if they happen to be 0/1 valued
PROTECTED = {"delirium_label", "subject_id", "hadm_id", "stay_id"}

binary_encoded = []
for col in df.columns:
    if col in PROTECTED:
        continue
    non_null = df[col].dropna().unique()
    if len(non_null) == 0:
        continue
    # subset-of-{0,1} test, tolerant of float storage (0.0 / 1.0)
    if not set(non_null).issubset({0, 1}):
        continue
    n_miss = int(df[col].isnull().sum())
    if n_miss == 0:
        continue  # already a clean binary column, leave as-is
    df[col] = df[col].fillna(2).astype(int)
    binary_encoded.append((col, n_miss))

print(f"\nBinary columns re-encoded as 0/1/2 (2 = unknown): {len(binary_encoded)}")
for col, n_miss in binary_encoded:
    print(f"  {col}: {n_miss} missing -> 2")
# ------------------------------------------------------------------
still_missing = df.isna().sum()
still_missing = still_missing[still_missing > 0].sort_values(ascending=False)
if len(still_missing):
    print(f"\nColumns still missing after encoding (NOT mode-filled): {len(still_missing)}")
    for col, n in still_missing.items():
        print(f"  {col}: {int(n)} ({100*n/len(df):.1f}%)")
else:
    print("\nNo remaining missingness.")

df.to_csv("cohort_all_features_capped.csv", index=False)
print("\nSaved: cohort_all_features_capped.csv")
```

## **4.5 · KNN Imputation** (continuous vitals/labs + discrete severity scores)

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
import time

df = pd.read_csv("cohort_all_features_capped.csv")
print(f"Loaded: {df.shape}")

# Build column list
SUFFIXES = ["_min", "_max", "_mean", "_last"]

PREFIXES = [
    "heart_rate", "sbp", "dbp", "mbp", "resp_rate", "temperature",
    "spo2", "glucose_vital", "hematocrit", "hemoglobin", "platelets",
    "wbc", "aniongap", "bicarbonate", "bun", "calcium", "chloride",
    "creatinine", "glucose_lab", "sodium", "potassium", "inr", "pt",
    "ptt", "mchc", "mcv", "rbc", "rdw",
]

STANDALONE = [
    "weight_admit", "urineoutput",
    "oasis_heartrate", "oasis_meanbp", "oasis_temp", "oasis_urineoutput",
    "rass_mean", "shock_index", "bun_cr_ratio",
]

# Vitals / labs (imputed by KNN in Stage 1).
knn_cols = []
for prefix in PREFIXES:
    for suffix in SUFFIXES:
        col = prefix + suffix
        if col in df.columns:
            knn_cols.append(col)

for col in STANDALONE:
    if col in df.columns:
        knn_cols.append(col)

# Discrete severity-score components (imputed by KNN in Stage 2, then rounded).
# Ordinal integer scores (RASS min/max/last, GCS motor/verbal/eyes, SOFA and
# LODS sub-scores, vw_score) that are not part of the vitals/labs list and are
# not binary, so no earlier step fills them.
SCORE_COLS = [
    "rass_min", "rass_max", "rass_last",
    "gcs_motor", "gcs_verbal", "gcs_eyes",
    "sofa_coag", "sofa_cardio", "sofa_renal",
    "lods_neurologic", "lods_cardiovascular", "lods_renal", "lods_hematologic",
    "vw_score",
]
SCORE_COLS = [c for c in SCORE_COLS if c in df.columns]


def round_half_up(frame):
    # Round to nearest integer with a 0.5 cutoff (0.5 -> 1, 2.5 -> 3).
    # np.round / pandas .round use banker's rounding (0.5 -> 0, 2.5 -> 2),
    # which is not the behavior we want here.
    return np.floor(frame.astype(float) + 0.5).astype(int)


def majority_whole(s, thresh=0.5):
    # True if the MAJORITY (> thresh) of a column's observed (non-null)
    # values are whole numbers. Run before imputation, so it reflects the
    # real recording granularity rather than KNN-produced fractions.
    v = s.dropna().values
    if len(v) == 0:
        return False
    return float(np.mean(np.isclose(v, np.round(v), atol=1e-9))) > thresh


# ------------------------------------------------------------------
# Detect which vitals/labs are recorded as whole numbers. Integer vitals
# and their min/max/last (heart rate, SpO2, resp rate, blood pressures) and
# integer labs (sodium, chloride, bicarbonate, BUN, glucose, platelets, MCV)
# are caught. Genuinely continuous labs (creatinine, temperature in C, INR,
# PT/PTT, hemoglobin, ratios) and most fractional *_mean aggregates are left
# as floats. These integer columns are rounded after imputation with the same
# 0.5 cutoff used for the scores.
#
# NOTE: a handful of lab *_mean columns are majority-whole (because those
# labs are often drawn once per window, so the "mean" is a single integer)
# and are therefore rounded too. To keep every mean aggregate continuous
# instead, exclude them here, e.g.:
#     int_vital_cols = [c for c in int_vital_cols if not c.endswith("_mean")]
# ------------------------------------------------------------------
int_vital_cols = [c for c in knn_cols if majority_whole(df[c])]
float_vital_cols = [c for c in knn_cols if c not in int_vital_cols]
print(f"\nVitals/labs kept as float (continuous): {len(float_vital_cols)}")
print(f"Vitals/labs rounded to int (majority whole): {len(int_vital_cols)}")
_mean_rounded = sorted(c for c in int_vital_cols if c.endswith("_mean"))
if _mean_rounded:
    print(f"  (includes {len(_mean_rounded)} *_mean columns: {_mean_rounded})")

# ------------------------------------------------------------------
# Stage 1: impute vitals / labs
# ------------------------------------------------------------------
knn_cols_missing = [c for c in knn_cols if df[c].isnull().any()]
print(f"\nStage 1 - vitals/labs: {len(knn_cols)} columns "
      f"({len(knn_cols_missing)} with missing values)")
print(f"Running KNN imputation (k=100)...")
t0 = time.time()
imputer = KNNImputer(n_neighbors=100)
df[knn_cols] = imputer.fit_transform(df[knn_cols])
print(f"Done in {time.time() - t0:.1f}s")

# ------------------------------------------------------------------
# Stage 2: impute discrete severity scores. The vitals/labs are complete
# after Stage 1, so here they only supply the distance space (their values
# are unchanged); only the score columns are filled.
# ------------------------------------------------------------------
score_missing = [c for c in SCORE_COLS if df[c].isnull().any()]
print(f"\nStage 2 - discrete scores: {len(SCORE_COLS)} columns "
      f"({len(score_missing)} with missing values)")
for c in score_missing:
    n = df[c].isnull().sum()
    print(f"    {c}: {n} missing ({100*n/len(df):.1f}%)")
print(f"Running KNN imputation (k=100)...")
t0 = time.time()
dist_cols = knn_cols + SCORE_COLS
df[dist_cols] = KNNImputer(n_neighbors=100).fit_transform(df[dist_cols])
print(f"Done in {time.time() - t0:.1f}s")

# ------------------------------------------------------------------
# Round every whole-number column with the 0.5 cutoff: the vitals/labs
# detected as majority-whole, plus all of the discrete scores.
# ------------------------------------------------------------------
round_cols = int_vital_cols + SCORE_COLS
still = df[round_cols].isnull().sum()
assert still.sum() == 0, f"Columns still missing after KNN: {still[still > 0].to_dict()}"
df[round_cols] = round_half_up(df[round_cols])
print(f"\nRounded to integers (0.5 cutoff, round half up): "
      f"{len(int_vital_cols)} vitals/labs + {len(SCORE_COLS)} scores")

remaining = int(df.isnull().sum().sum())
print(f"Remaining missing (whole frame): {remaining}")

df.to_csv("cohort_all_features_capped.csv", index=False)
print("Saved: cohort_all_features_capped.csv")
```

## **4.6 · Split Columns into min/ max / mean/ last**

```python
from collections import defaultdict
df = pd.read_csv("cohort_all_features_capped.csv")
print(f"Loaded: {df.shape}")

# Find prefixes that have all 4 suffixes
SUFFIXES = ["_min", "_max", "_mean", "_last"]
prefix_map = defaultdict(set)

for c in df.columns:
    for s in SUFFIXES:
        if c.endswith(s):
            prefix = c[:-len(s)]
            prefix_map[prefix].add(s)

# Only split columns with all 4 variants
split_prefixes = sorted(p for p in prefix_map if len(prefix_map[p]) == 4)

# Everything else is static (same across all files)
all_variable = set()
for p in split_prefixes:
    for s in SUFFIXES:
        all_variable.add(p + s)

static_cols = [c for c in df.columns if c not in all_variable]

print(f"Static columns: {len(static_cols)}")
print(f"Variable prefixes (split): {len(split_prefixes)}")

# Build and save each file
for suffix in SUFFIXES:
    label = suffix[1:]  # min, max, mean, last

    # Start with static columns
    out = df[static_cols].copy()

    # Add variable columns with their original names
    var_cols = [prefix + suffix for prefix in split_prefixes]
    for col in var_cols:
        out[col] = df[col].values

    fname = f"cohort_{label}.csv"
    out.to_csv(fname, index=False)
    print(f"  {fname}: {out.shape}")

print("Done")
```
